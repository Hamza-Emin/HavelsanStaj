```python
import spacy

nlp = spacy.load("tr_core_news_md")
doc = nlp("Vodafone ile Turkcell arasında karşılaştırma yapmak gerekirse, Vodafone hız konusunda daha iyiyken Turkcell daha ucuz.")

print("{:25} {:10} {:10} {:20} {:10} {:10} {:50} {:50}".format('Token','Pos-Tag','Dep','Head','Pos-Head','NER', 'Lefts', 'Rights'))
print("_" * 150)
for token in doc:
    print("{:2} {:10} {:10} {:20} {:10} {:10} {:50} {:50}".format(token.text, token.pos_ , token.dep_, token.head.text, token.head.pos_, token.ent_type_,  "-".join([w.text for w in token.lefts]), "-".join([w.text for w in token.rights])))
```

```python
from spacy import displacy
displacy.render(doc, style='dep', jupyter=True)
```
